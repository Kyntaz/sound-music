# Make Spectogram

```python
%matplotlib inline
import librosa as lr
import librosa.display
import numpy as np
import matplotlib.pyplot as plt

path = "../output/complement_splash.wav"
wave, sr = lr.load(path)
fft = lr.stft(wave)
fft_a = lr.amplitude_to_db(np.abs(fft), ref=np.max)
plt.figure(figsize=[9, 4])
lr.display.specshow(fft_a, x_axis="time", y_axis="log")
plt.colorbar(format='%+2.0f dB')
```
